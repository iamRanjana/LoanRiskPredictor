```python
# Checking  the mean value in AMT_REQ_CREDIT_BUREAU_HOUR COLUMN.

inp_0.AMT_REQ_CREDIT_BUREAU_HOUR.mean()
```

# Handling the outliers

```python
#finding outliers in each required columns
# i choose these columns upon understand the column description read from the column description file given.
amtcols = [x for x in list(inp_0.columns) if 'AMT' in x]
for i in amtcols:
    sns.boxplot(inp_0[i])
    plt.xlabel(i)
    plt.show()
```

```python
# replacing the outliers
for i in amtcols:
    p95 = inp_0[i].median()
    inp_0.loc[inp_0[i] == inp_0[i].max(), i] = p95
```

```python
#converting days to years and dropping the days columns by adding the years columns
days=['DAYS_BIRTH','DAYS_EMPLOYED','DAYS_REGISTRATION','DAYS_ID_PUBLISH']
years=['YEARS_BIRTH','YEARS_EMPLOYED','YEARS_REGISTRATION','YEARS_ID_PUBLISH']
for i,j in zip(days,years):
    inp_0[j]=round(inp_0[i]/365,2)
```

```python
# replacing the outlier with median because in this column even after replacing with 95th percentile the value is around 1000 experiance of a person being 1000 is practically impossible.
# replacing only with median because the mean is effected by the outlier and median is better compared to mode.
inp_0.loc[inp_0['YEARS_EMPLOYED'] == inp_0['YEARS_EMPLOYED'].max(),'YEARS_EMPLOYED'] =inp_0['YEARS_EMPLOYED'].median()
```
